Image manipulation

```python
# Import libraries
import os
import PIL
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np

from skimage import io, transform
from skimage.transform import resize
```

```python
### Open image and convert to grayscale.

# Placeholder--you need to save the grayscale image in the 'img' variable
img = None

# >>> open test image <<<
image_path = "../data/raw/electronic-components-bmp/capacitor/0.bmp"
img = Image.open(image_path)
img.show()
```

```python

# >>> convert to grayscale image <<<
image_path = "../data/raw/electronic-components-bmp/capacitor/0.bmp"
img = img.convert("L")

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
### Convert the Image object to a Numpy array

# Save Numpy array in the 'img_array' variable
img_array = None

img_array = np.array(img)
print(img_array.shape)
print(img_array)

plt.imshow(img_array, cmap='gray', vmin=0, vmax=255)
```

```python
# Resize image to a 28x28 image

img_resized = None
img_resized = transform.resize(img_array, (28, 28), anti_aliasing = True)
print(img_resized)

# transform.resize stores data in float. Rescaling to uint8 is necessary.
img_resized_uint8 = (img_resized * 255).astype(np.uint8)
print(img_resized_uint8)
plt.imshow(img_resized_uint8, cmap='gray', vmin=0, vmax=255)
```

Draw on the image
### Change the required elements in the Numpy array so that a 10x10 white square is drawn in the 
# middle.

```python
# Start with the resized image
img_with_box = np.copy(img_resized_uint8) # copy resized image into new variable

plt.imshow(img_resized_uint8, cmap='gray', vmin=0, vmax=255)

box_size = 10
white_box = np.ones((box_size, box_size), dtype=np.uint8) * 255 # 255: white, 0: black
row, col = 5, 8 # choose arbitrary location of the white box
img_with_box[row:row+box_size, col:col+box_size] = white_box

plt.imshow(img_with_box, cmap='gray', vmin=0, vmax=255)
```
